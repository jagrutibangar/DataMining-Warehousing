```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum() #no Duplicates
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='inferno')
```

```python
df['Ca'].fillna(df['Ca'].mean())  #fill the nan values with mean
```

```python
df['Thal'].fillna(df['Thal'].mode()[0])   #fill nan values with mode
```

```python
df.isnull().sum()
```

```python
plt.figure()
sns.boxplot(y=df['RestBP'])
plt.title("Boxplot for RestBP")
plt.show()
```

```python
Q1 = df['RestBP'].quantile(0.25)
Q3 = df['RestBP'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers = df[(df['RestBP'] < lower_bound) | (df['RestBP'] > upper_bound)]
outliers
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Chol'], kde=True)
plt.title('Distribution of Chol')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,2)
sns.histplot(df['RestBP'], kde=True)
```
